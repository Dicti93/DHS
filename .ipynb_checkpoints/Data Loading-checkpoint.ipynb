```python
import os
import pandas as pd
```

```python
# Base directories
base_dirs = {
    'Data Wave 2022': '2022',
    'Data Wave 2023': '2023'
}

# Dictionary to store DataFrames
dataframes = {}
```

```python
# Print the current working directory
current_working_directory = os.getcwd()
print(f"Current Working Directory: {current_working_directory}")

# List all files and directories in the current working directory
print("Files and directories in the current working directory:")
for item in os.listdir(current_working_directory):
    print(item)
```

```python
# Verify that base directories exist
for base_dir in base_dirs.keys():
    full_path = os.path.join(current_working_directory, base_dir)
    if not os.path.exists(full_path):
        print(f"Base directory does not exist: {full_path}")
    else:
        print(f"Found base directory: {full_path}")
```

```python
# Function to process .sav files
def process_sav_file(file_path, folder_name, year):
    try:
        data = pd.read_spss(file_path)
        print(f"Successfully read {file_path}")
        # Create a key for the DataFrame dictionary
        key = f"{folder_name} {year}"
        dataframes[key] = data
    except Exception as e:
        print(f"Error reading {file_path}: {e}")
```

```python
# Iterate through the base directories
for base_dir, year in base_dirs.items():
    full_path = os.path.join(current_working_directory, base_dir)
    for root, dirs, files in os.walk(full_path):
        print(f"Exploring directory: {root}")  # Debugging output
        for file in files:
            if file.endswith('.sav'):
                folder_name = os.path.basename(root)
                file_path = os.path.join(root, file)
                print(f"Found .sav file: {file_path}")  # Debugging output
                process_sav_file(file_path, folder_name, year)
```

```python
# Check the contents of the dataframes dictionary
print("DataFrames keys:", list(dataframes.keys()))  # Debugging output
```

```python
# print the first few rows of each DataFrame
for key, df in dataframes.items():
    print(f"DataFrame for {key}:")
    print(df.head())
    print("\n")
```

## Spatial Representation of Survey Respondents

```python
hh_df_spatial = dataframes['hhi2022en_spss_1.0 2022'][["nohhold", "nomem", "prov"]]
hh_spatial_agg = hh_df_spatial[["prov"]].groupby(by = "prov").size().reset_index()
hh_spatial_agg.columns = ["prov", "count"]
hh_spatial_agg.sort_values(by = "count", ascending = False)
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

gdf = gpd.read_file('nl.json')
gdf = gdf.rename(columns={'name': 'prov'})
merged = gdf.set_index('prov').join(hh_spatial_agg.set_index('prov'))

# Plot the map with a color gradient representing the number of respondents
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
merged.plot(column='count', ax=ax, legend=True, cmap='OrRd', 
            legend_kwds={'label': "Number of Respondents",
                         'orientation': "horizontal"})

# Set the title and show the plot
plt.title('Survey Respondents per Province in the Netherlands')
plt.show()
```

Dark red on the left is Zuid-Holland which encompasses large cities like Rotterdam, The Hague and Leiden. Above is Noord-Holland with Amsterdam. The two other most frequent regions are Noord-Brabant and Gelderland.

## EDA of a merged dataset

```python
# Convert column names to lowercase
dataframes['inc2022en_spss_1.0 2022'].columns = dataframes['inc2022en_spss_1.0 2022'].columns.str.lower()
dataframes['psy2022en_spss_1.0 2022'].columns = dataframes['psy2022en_spss_1.0 2022'].columns.str.lower()
dataframes['hhi2022en_spss_1.0 2022'].columns = dataframes['hhi2022en_spss_1.0 2022'].columns.str.lower()

hh_df = dataframes['hhi2022en_spss_1.0 2022'][["nohhold", "nomem", "gebjaar", "geslacht", "oplmet", "prov"]] 
hh_df = hh_df.dropna(subset=['prov'])

inc_df = dataframes['inc2022en_spss_1.0 2022'][["nohhold", "nomem","gez3", "gez5", "gez7", "ij161"]]
inc_df = inc_df.dropna(subset=['ij161'])
inc_df['ij161'] = pd.to_numeric(inc_df['ij161'], errors='coerce')

psy_df = dataframes['psy2022en_spss_1.0 2022'][["nohhold", "nomem", "finsitu", "ecsit", "opzij"]]
psy_df = psy_df.dropna(subset=['finsitu', 'ecsit', 'opzij'])

merged_df = hh_df.merge(inc_df, on=['nohhold', 'nomem']).merge(psy_df, on=['nohhold', 'nomem'])
```

## Spatial Comparison of Income and Financial Situation

```python
spatial_df = merged_df[["prov", "gebjaar", "gez3", "gez5", "gez7", "ij161", "finsitu", "ecsit", "opzij"]].copy()
spatial_df = spatial_df.applymap(lambda s: s.lower() if type(s) == str else s)

category_order = ['poor', 'not so good', 'fair', 'good', 'excellent']
spatial_df['gez3'] = pd.Categorical(spatial_df['gez3'], categories=category_order, ordered=True)
spatial_df['gez3'] = spatial_df['gez3'].cat.codes

yes_no_mapping = {'yes': 1, 'no': 0}
spatial_df['gez5'] = spatial_df['gez5'].map(yes_no_mapping)
spatial_df['opzij'] = spatial_df['opzij'].map(yes_no_mapping)

category_order = ['no', 'yes, every now and then', 'yes, i smoke every day']
spatial_df['gez7'] = pd.Categorical(spatial_df['gez7'], categories=category_order, ordered=True)
spatial_df['gez7'] = spatial_df['gez7'].cat.codes

category_order = ['there are debts', 'need to draw upon savings', 'it is just about manageable', 'some money is saved' , 'a lot of money can be saved']
spatial_df['finsitu'] = pd.Categorical(spatial_df['finsitu'], categories=category_order, ordered=True)
spatial_df['finsitu'] = spatial_df['finsitu'].cat.codes

category_order = ['much worse', 'worse', '(about) the same', 'better',  'much better']
spatial_df['ecsit'] = pd.Categorical(spatial_df['ecsit'], categories=category_order, ordered=True)
spatial_df['ecsit'] = spatial_df['ecsit'].cat.codes

spatial_group = spatial_df.groupby(by =["prov"]).mean().reset_index()
```

```python
gdf = gdf.applymap(lambda s: s.lower() if type(s) == str else s)
merged = gdf.set_index('prov').join(spatial_group.set_index('prov'))
```

```python
# Plot the map with a color gradient representing the number of respondents
fig, ax = plt.subplots(1, 2, figsize=(15, 10))
merged.plot(column='ij161', ax=ax[0], legend=True, cmap='OrRd', 
            legend_kwds={'label': "Income",
                         'orientation': "horizontal"})
merged.plot(column='finsitu', ax=ax[1], legend=True, cmap='OrRd', 
            legend_kwds={'label': "Financial Situation",
                         'orientation': "horizontal"})

# Set the title and show the plot
ax[0].title.set_text('Income Average per Province in the Netherlands')
ax[1].title.set_text('Self-Assessed Financial Situation per Province in the Netherlands')
plt.show()
```

## Age Segments Comparison

```python
age_df = spatial_df[["gebjaar", "gez3", "gez5", "gez7", "ij161", "finsitu", "ecsit", "opzij"]].copy()
age_df["age"] = 2022- age_df["gebjaar"]
age_df = age_df[["age", "gez3", "gez5", "gez7", "ij161", "finsitu", "ecsit", "opzij"]]

bins = [18, 30, 40, 50, 60, 100]
labels = ['18-30', '30-40', '40-50', '50-60', '60+']

# Use pd.cut to categorize the ages
age_df['age_group'] = pd.cut(age_df['age'], bins=bins, labels=labels, right=False)
age_df = age_df[["age_group", "gez3", "gez5", "gez7", "ij161", "finsitu", "ecsit", "opzij"]]
age_df.columns = ["age_group", "health_status", "longterm_illness", "smoking", "household_income", "financial_situation", "economic_situation", "savings_behavior"]

age_df.groupby(by =["age_group"]).mean().reset_index()
```

```python
import seaborn as sns

# Set up the Seaborn FacetGrid for faceted bar plots
g = sns.FacetGrid(age_df.melt(id_vars='age_group'), col='variable', col_wrap=4, sharey=False, hue='age_group', palette='pastel')
g.map(sns.barplot, 'age_group', 'value', order=['18-30', '30-40', '40-50', '50-60', '60+'], errorbar = None)

# Adjust the plots for better readability
g.set_titles("{col_name}")
g.set_axis_labels("", "Value")
for ax in g.axes.flat:
    for label in ax.get_xticklabels():
        label.set_rotation(90)
        
plt.legend(title='Age Group', bbox_to_anchor=(1.05, 1), loc='best')

# Show the plot
plt.suptitle('Comparison of Health and Financial Situation between Age Groups')
plt.tight_layout()
plt.show()
```
