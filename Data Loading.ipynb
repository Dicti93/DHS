```python
import os
import pandas as pd
```

```python
# Base directories
base_dirs = {
    'Data Wave 2022': '2022',
    'Data Wave 2023': '2023'
}

# Dictionary to store DataFrames
dataframes = {}
```

```python
# Print the current working directory
current_working_directory = os.getcwd()
print(f"Current Working Directory: {current_working_directory}")

# List all files and directories in the current working directory
print("Files and directories in the current working directory:")
for item in os.listdir(current_working_directory):
    print(item)
```

```python
# Verify that base directories exist
for base_dir in base_dirs.keys():
    full_path = os.path.join(current_working_directory, base_dir)
    if not os.path.exists(full_path):
        print(f"Base directory does not exist: {full_path}")
    else:
        print(f"Found base directory: {full_path}")
```

```python
# Function to process .sav files
def process_sav_file(file_path, folder_name, year):
    try:
        data = pd.read_spss(file_path)
        print(f"Successfully read {file_path}")
        # Create a key for the DataFrame dictionary
        key = f"{folder_name} {year}"
        dataframes[key] = data
    except Exception as e:
        print(f"Error reading {file_path}: {e}")
```

```python
# Iterate through the base directories
for base_dir, year in base_dirs.items():
    full_path = os.path.join(current_working_directory, base_dir)
    for root, dirs, files in os.walk(full_path):
        print(f"Exploring directory: {root}")  # Debugging output
        for file in files:
            if file.endswith('.sav'):
                folder_name = os.path.basename(root)
                file_path = os.path.join(root, file)
                print(f"Found .sav file: {file_path}")  # Debugging output
                process_sav_file(file_path, folder_name, year)
```

```python
# Check the contents of the dataframes dictionary
print("DataFrames keys:", list(dataframes.keys()))  # Debugging output
```

```python
# print the first few rows of each DataFrame
for key, df in dataframes.items():
    print(f"DataFrame for {key}:")
    print(df.head())
    print("\n")
```
